```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Define attack types
attack_types = ['Normal', 'Flooding', 'Blackhole', 'Grayhole', 'Scheduling']

# Training Confusion Matrix
training_cm = np.array([
    [272058, 0, 13, 0, 0],
    [0, 271705, 0, 0, 0],
    [49, 0, 271861, 0, 0],
    [5, 240, 196, 271981, 87],
    [0, 0, 0, 88, 271981]
])

# Testing Confusion Matrix
testing_cm = np.array([
    [67991, 0, 4, 0, 0],
    [0, 68360, 0, 1, 0],
    [36, 0, 68120, 0, 0],
    [0, 65, 79, 67358, 55],
    [1, 0, 0, 145, 67851]
])

# Plot Training Confusion Matrix
plt.figure(figsize=(10, 8))
sns.heatmap(training_cm, annot=True, fmt='d', cmap='Blues', xticklabels=attack_types, yticklabels=attack_types)
plt.title('Confusion Matrix - Training (OOS-GNN)')
plt.xlabel('Predicted Attack Type')
plt.ylabel('Actual Attack Type')
plt.show()

# Plot Testing Confusion Matrix
plt.figure(figsize=(10, 8))
sns.heatmap(testing_cm, annot=True, fmt='d', cmap='Greens', xticklabels=attack_types, yticklabels=attack_types)
plt.title('Confusion Matrix - Testing (OOS-GNN)')
plt.xlabel('Predicted Attack Type')
plt.ylabel('Actual Attack Type')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Metrics
metrics = ['A', 'Sn', 'Sp', 'Fscore', 'AUC']

# Values
training_scores = [99.77, 97.96, 99.73, 94.56, 98.84]
testing_scores = [99.60, 98.064, 99.662, 94.72, 99.09]

# Bar positions
x = np.arange(len(metrics))
width = 0.35  # Width of the bars

# Plotting the bar chart
plt.figure(figsize=(10, 6))
plt.bar(x - width/2, training_scores, width, label='Training', color='purple')
plt.bar(x + width/2, testing_scores, width, label='Testing', color='lightgreen')

# Labels and titles
plt.xlabel('Performance Metrics', fontsize=12)
plt.ylabel('Percentage (%)', fontsize=12)
plt.title('Comparison of Performance Metrics', fontsize=14)
plt.xticks(x, metrics)
plt.ylim(90, 100)  # Adjust based on the range of your values
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Models
models = ['OOA-GNN', 'BCOA-MLID', 'AdaBoost', 'GB Model', 'XGBoost', 'KNN-AOA', 'KNN-PSO']

# Performance metrics
accuracy = [99.68, 99.63, 95.69, 94.58, 96.83, 97.20, 92.89]
sensitivity = [98.01, 97.91, 95.77, 95.25, 96.10, 96.49, 95.63]
specificity = [99.73, 99.67, 95.00, 94.09, 94.43, 96.34, 95.08]
fscore = [94.56, 94.52, 90.31, 93.31, 91.52, 90.23, 92.99]

# Bar width
bar_width = 0.18
index = np.arange(len(models))

# Plotting the bars
plt.figure(figsize=(16, 8))
plt.bar(index, accuracy, bar_width, label='Accuracy (A)', color='#4CAF50')
plt.bar(index + bar_width, sensitivity, bar_width, label='Sensitivity (Sn)', color='#2196F3')
plt.bar(index + 2 * bar_width, specificity, bar_width, label='Specificity (Sp)', color='#FFC107')
plt.bar(index + 3 * bar_width, fscore, bar_width, label='F-score', color='#E91E63')

# Labels and title
plt.xlabel('Models/Methods', fontsize=14)
plt.ylabel('Percentage (%)', fontsize=14)
plt.title('Comparison of Different Models Based on Performance Metrics', fontsize=16)
plt.xticks(index + 1.5 * bar_width, models, rotation=30)
plt.ylim(85, 100)
plt.legend(loc='upper left', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Display the chart
plt.tight_layout()
plt.show()
```
